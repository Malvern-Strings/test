```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from numpy import polyfit,poly1d
import xlrd
```

```python
file_location='E:\Electronic Technology Experiment\实验七 半导体器件传输特性研究\semiconductor.xls'
data=xlrd.open_workbook(file_location)
sheet=data.sheet_by_index(0)  
```

```python
#1N4007
vol=np.array([sheet.cell_value(2,c) for c in range(2,22)])
cur=np.array([sheet.cell_value(3,c) for c in range(2,22)])

# print(vol)
# print(cur)

fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111)

ax.scatter(vol[:],cur[:],)
ax.plot(vol[:],cur[:])

ax.set_xlabel(r'$U/V$',fontsize=13)
ax.set_ylabel(r'$I/mA$',fontsize=13)

#ax.legend(frameon=True,fontsize=13)

plt.grid()
plt.show()
```

```python
#red LED&blue LED
volr=np.array([sheet.cell_value(10,c) for c in range(2,18)])
curr=np.array([sheet.cell_value(11,c) for c in range(2,18)])
volb=np.array([sheet.cell_value(14,c) for c in range(2,19)])
curb=np.array([sheet.cell_value(15,c) for c in range(2,19)])

# print(volr)
# print(curr)
# print(volb)
# print(curb)

fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111)

ax.scatter(volr[:],curr[:],c='r')
ax.plot(volr[:],curr[:],c='r',label='red LED')
ax.scatter(volb[:],curb[:],c='b')
ax.plot(volb[:],curb[:],c='b',label='blue LED')

ax.set_xlabel(r'$U/V$',fontsize=13)
ax.set_ylabel(r'$I/mA$',fontsize=13)

ax.legend(frameon=True,fontsize=13)

plt.grid()
plt.show()
```

```python
#NPN输入特性
vol00=np.array([sheet.cell_value(19,c) for c in range(2,11)])
cur00=np.array([sheet.cell_value(20,c) for c in range(2,11)])
vol07=np.array([sheet.cell_value(22,c) for c in range(2,11)])
cur07=np.array([sheet.cell_value(23,c) for c in range(2,11)])
vol10=np.array([sheet.cell_value(25,c) for c in range(2,11)])
cur10=np.array([sheet.cell_value(26,c) for c in range(2,11)])


vol07=np.append(vol07,1.103)
cur07=np.append(cur07,40)
vol10=np.append(vol10,1.103)
cur10=np.append(cur10,43)

fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111)

ax.scatter(vol00[:],cur00[:],)
ax.plot(vol00[:],cur00[:],label=r'$U_{CE}=0V$')
ax.scatter(vol07[:],cur07[:],)
ax.plot(vol07[:],cur07[:],label=r'$U_{CE}=1.0AV$')
ax.scatter(vol10[:],cur10[:],)
ax.plot(vol10[:],cur10[:],label=r'$U_{CE}=0.7V$')

ax.set_xlabel(r'$u_{BE}/V$',fontsize=13)
ax.set_ylabel(r'$i_B/mA$',fontsize=13)

ax.legend(frameon=True,fontsize=13)

plt.grid()
plt.show()

```

```python
#NPN输出特性
vol15=np.array([sheet.cell_value(30,c) for c in range(3,23)])
cur15=np.array([sheet.cell_value(31,c) for c in range(3,23)])
vol30=np.array([sheet.cell_value(34,c) for c in range(3,23)])
cur30=np.array([sheet.cell_value(35,c) for c in range(3,23)])

# print(vol15)
# print(cur15)
# print(vol15)
# print(cur30)

fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111)

ax.scatter(vol15[:],cur15[:],)
ax.plot(vol15[:],cur15[:],label=r'$I_B=1.5mA$')
ax.scatter(vol30[:],cur30[:],)
ax.plot(vol30[:],cur30[:],label=r'$I_B=3.0mA$')

ax.set_xlabel(r'$u_{CE}/V$',fontsize=13)
ax.set_ylabel(r'$i_C/mA$',fontsize=13)

ax.legend(frameon=True,fontsize=13)

plt.grid()
plt.show()
```
